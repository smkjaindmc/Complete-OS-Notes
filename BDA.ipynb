```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import folium
from folium import plugins
plt.rcParams['figure.figsize']=10,12
import warnings
warnings.filterwarnings('ignore')
```

```
df=pd.read_csv('/content/covid_india.csv')
df_india=df.copy()
df
```

```
India_coord = pd.read_excel('/content/Indian Coordinates.xlsx')
India_coord
```

```
total_cases = df['Total Confirmed cases'].sum()
print('Total number of confirmed cases in India (as of 27 Nov,2020): ',total_cases)
```

```
df.style.background_gradient(cmap='Reds')
```

```
dp_full = pd.merge(India_coord,df,on='Name of State / UT')
map = folium.Map(location = [20,70], zoom_start=4, tiles='Stamenterrain')

for lat,lon,value,name in zip(dp_full['Latitude'],dp_full['Longitude'],dp_full['Total Confirmed cases'],dp_full['Name of State / UT']):
  folium.CircleMarker([lat,lon], radius=value*0.00008,color='red', popup=('<strong>State</strong>: ' + str(name).capitalize()), fill=True).add_to(map)
map
```

```
f, ax = plt.subplots(figsize=(12,8))
data = dp_full[['Name of State / UT','Total Confirmed cases','Cured/Discharged/Migrated','Deaths']]
data.sort_values('Total Confirmed cases',ascending=False,inplace=True)
sns.set_color_codes("pastel")
sns.barplot(x="Total Confirmed cases", y="Name of State / UT",data=data, label="Total",color="r")
sns.set_color_codes("muted")
sns.barplot(x="Cured/Discharged/Migrated", y="Name of State / UT",data=data, label="Cured",color="g")
```
